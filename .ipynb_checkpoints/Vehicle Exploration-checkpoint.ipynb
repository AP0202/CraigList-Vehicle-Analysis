```python
#Data Manipulation
```

```python
df2 = df.head(100)
df2.rename(index = str , columns={'region':'new_region'},inplace = False)
```

```python
url_region_price = df[['url','region','price']]
print(url_region_price)
```

```python
data_no_url = df.drop('url',axis=1)#Dropping column asix = 1 for column and axis=0 for row
data_no_url.head(10)
```

```python
df['age'] = 2020 - df['year']#column Creation
df['current_year'] = 2019
df[['age','current_year']].head(20)
```

```python
#new car age < 10 years filtering by this criteria
data_new_cars = df[df.age <=10]
data_new_cars['age']
```

```python
#New Feature Creation
df['price_per_mile'] = df['price'] / df['odometer']
df.head(5)
```

```python
#Creating Function.
def price2x(x):
    return x*2

df['price2'] = df['price'].apply(price2x)
display(df.price.head(10))
display(df.price2.head(10))
```

```python
#Lambda Function
df['price3'] = df['price'].apply(lambda x: x*3)
display(df.price3.head(10))
```

```python
df['isexpensive'] = df['price'].apply(lambda x: 'expensive' if x>10000 else 'cheap')
display(df['isexpensive'].head(10))
```

```python
df['new_and_cheap'] = df.apply(lambda x: 'yes'\
                                       if x['price'] < 10000 \
                                       and x['age'] < 10 \
                                       else 'no',axis=1)
display(df['new_and_cheap'].head(10))
```

```python
#Dummy Variable
dummy_vars = pd.get_dummies(df[['price','year','fuel','transmission','type']])
dummy_vars.head(100)
```

```python
#Pivot Tables
pd.pivot_table(df , index = 'year' , columns = 'type',\
              values = 'price', aggfunc='mean').sort_index(ascending = False)
```

```python
pd.pivot_table(df , index = 'year' , columns = 'type',\
              values = 'price', aggfunc='count').sort_index(ascending = False).plot()
```

```python
#groupby
grouped_data = df.groupby(['type','year'], as_index = False).mean()
display(grouped_data)
```

```python
#Merging
data1 = df[['url','region']]
data2 = df[['url','price']]
df_joined = pd.merge(data1 , data2 , on = 'url')
display(df_joined)
```

```python
#Appending
sample1 = df.sample(100 , random_state = 1)
sample2 = df.sample(100 , random_state = 2)
sample1.append(sample2)
```

```python
#Writing manipulated data to csv
df.head(1000).to_csv('top1000.csv')
```
